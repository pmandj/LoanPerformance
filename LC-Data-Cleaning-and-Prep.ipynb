```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load all quarters 
q1 = pd.read_csv('LoanStats_2017Q1.csv.zip', skiprows=1, compression='zip', low_memory=False)
q2 = pd.read_csv('LoanStats_2017Q2.csv.zip', skiprows=1, compression='zip', low_memory=False)
q3 = pd.read_csv('LoanStats_2017Q3.csv.zip', skiprows=1, compression='zip', low_memory=False)
q4 = pd.read_csv('LoanStats_2017Q4.csv.zip', skiprows=1, compression='zip', low_memory=False)
#combine the quarters into a single dataset
df = pd.concat([q1,q2,q3,q4])
```

```python
df.head()
```

```python
#renaming columns
df.rename(columns={'installment':'monthly_pmt'}, inplace=True)
df.head()
```

```python
list(df.columns)
```

```python
#select only the columns you need from the data set and drop null vals
new_df=df.loc[:,['loan_amnt','term','int_rate','monthly_pmt','grade','issue_d','purpose', 'loan_status','dti','recoveries',
                'chargeoff_within_12_mths','all_util','tax_liens','pub_rec_bankruptcies','addr_state','delinq_2yrs',
                'delinq_amnt','annual_inc','num_tl_90g_dpd_24m','num_tl_30dpd']].dropna()

new_df.head()
```

```python
#remove the % sign and the term 'months' from int_rate and term columns
#change the data type for each of the columns
new_df['term'] = new_df.term.str.replace('months', '').astype(int)
new_df['int_rate'] = new_df.int_rate.str.replace('%','').astype(float)
new_df['issue_d'] = new_df.issue_d.str.replace('-2017','')
```

```python
#create a field that dislays ROI
new_df['total_return'] = new_df['term']*new_df['monthly_pmt']
new_df['roi'] = (new_df['total_return']/new_df['loan_amnt'])*100
new_df.head()
```
